```python
import nltk
import pandas as pd
from string import punctuation
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize 
from nltk.stem import PorterStemmer
```

```python
f=open(("001.txt"),'r')
text=f.read()
f.close()
print(text)
```

```python
sent_tokens = sent_tokenize(text)
word_tokens = word_tokenize(text)
```

```python
sent_tokens
```

```python
word_tokens
```

```python
word_tokens_lower = [word.lower() for word in word_tokens]
```

```python
stopWords=list(set(stopwords.words('english')))
word_tokens_refined=[word for word in word_tokens_lower if word not in stopWords]
```

```python
print(len(word_tokens))
```

```python
print(len(word_tokens_refined))
```

```python
ps=PorterStemmer()
stem=[]
for word in word_tokens_refined:
    stem.append(ps.stem(word))
    
word_tokens_refined=stem
```

```python
FreqTable={}
for word in word_tokens_refined:
    if word in FreqTable:
        FreqTable[word]+=1
    else:
        FreqTable[word]=1
print(len(FreqTable))
```

```python
FreqTable
```

```python
maxfreq=max(FreqTable.values())
print(maxfreq)
```

```python
sentence_scores = {}
for sent in sent_tokens:
    sentence = sent.split(" ")
    for word in sentence:        
        if word.lower() in FreqTable.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = FreqTable[word.lower()]
            else:
                sentence_scores[sent] += FreqTable[word.lower()]
```

```python
sum = 0
for sent in sentence_scores:
    sum+=sentence_scores[sent]
average=int(sum/len(sentence_scores))
print(average)
```

```python
summary=''
for sent in sent_tokens:
    if(sentence_scores[sent]>1.2*average):
        summary+=""+sent
print(summary)
```
